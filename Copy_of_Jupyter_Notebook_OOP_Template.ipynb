**Procedure:**

1.   A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

```python
def compute_class_standing(hoa, quiz, assignment):
    """
    Computes Class Standing (CS)

    Parameters:
    hoa (float): Hands-on Activities average
    quiz (float): Quiz average
    assignment (float): Assignment average

    Returns:
    float: Class Standing grade
    """
    return (0.50 * hoa) + (0.30 * quiz) + (0.20 * assignment)


# -------- PRELIM --------
prelim_exam = float(input("Enter Prelim Exam grade: "))
prelim_hoa = float(input("Enter Prelim HOA average: "))
prelim_quiz = float(input("Enter Prelim Quiz average: "))
prelim_assignment = float(input("Enter Prelim Assignment average: "))

prelim_cs = compute_class_standing(prelim_hoa, prelim_quiz, prelim_assignment)
prelim_grade = (0.50 * prelim_exam) + (0.50 * prelim_cs)


# -------- MIDTERM --------
midterm_exam = float(input("\nEnter Midterm Exam grade: "))
midterm_hoa = float(input("Enter Midterm HOA average: "))
midterm_quiz = float(input("Enter Midterm Quiz average: "))
midterm_assignment = float(input("Enter Midterm Assignment average: "))

midterm_cs = compute_class_standing(midterm_hoa, midterm_quiz, midterm_assignment)
midterm_current = (0.50 * midterm_exam) + (0.50 * midterm_cs)
midterm_grade = (1/3 * prelim_grade) + (2/3 * midterm_current)


# -------- FINAL --------
final_exam = float(input("\nEnter Final Exam grade: "))
final_hoa = float(input("Enter Final HOA average: "))
final_quiz = float(input("Enter Final Quiz average: "))
final_assignment = float(input("Enter Final Assignment average: "))

final_cs = compute_class_standing(final_hoa, final_quiz, final_assignment)
final_current = (0.50 * final_exam) + (0.50 * final_cs)
final_grade = (1/3 * midterm_grade) + (2/3 * final_current)


# -------- OUTPUT --------
print("\n===== GRADE SUMMARY =====")
print(f"Prelim Grade: {prelim_grade:.2f}")
print(f"Midterm Grade: {midterm_grade:.2f}")
print(f"Final Grade: {final_grade:.2f}")
```

```python
def compute_class_standing(hoa, quiz, assignment):
    """Computes class standing (CS)."""
    return (0.50 * hoa) + (0.30 * quiz) + (0.20 * assignment)


def tips_numerical_grade(final_grade):
    """Converts final grade to T.I.P.s numerical grade."""
    if final_grade >= 96:
        return 1.00
    elif final_grade >= 94:
        return 1.25
    elif final_grade >= 92:
        return 1.50
    elif final_grade >= 89:
        return 1.75
    elif final_grade >= 87:
        return 2.00
    elif final_grade >= 84:
        return 2.25
    elif final_grade >= 82:
        return 2.50
    elif final_grade >= 79:
        return 2.75
    elif final_grade >= 75:
        return 3.00
    else:
        return 5.00


for i in range(1, 4):
    print(f"\n===== STUDENT {i} =====")
    name = input("Enter student name: ")

    # PRELIM
    p_exam = float(input("Prelim Exam: "))
    p_hoa = float(input("Prelim HOA: "))
    p_quiz = float(input("Prelim Quiz: "))
    p_assign = float(input("Prelim Assignment: "))

    prelim_cs = compute_class_standing(p_hoa, p_quiz, p_assign)
    prelim_grade = 0.50 * p_exam + 0.50 * prelim_cs

    # MIDTERM
    m_exam = float(input("\nMidterm Exam: "))
    m_hoa = float(input("Midterm HOA: "))
    m_quiz = float(input("Midterm Quiz: "))
    m_assign = float(input("Midterm Assignment: "))

    midterm_cs = compute_class_standing(m_hoa, m_quiz, m_assign)
    midterm_current = 0.50 * m_exam + 0.50 * midterm_cs
    midterm_grade = (1/3 * prelim_grade) + (2/3 * midterm_current)

    # FINAL
    f_exam = float(input("\nFinal Exam: "))
    f_hoa = float(input("Final HOA: "))
    f_quiz = float(input("Final Quiz: "))
    f_assign = float(input("Final Assignment: "))

    final_cs = compute_class_standing(f_hoa, f_quiz, f_assign)
    final_current = 0.50 * f_exam + 0.50 * final_cs
    final_grade = (1/3 * midterm_grade) + (2/3 * final_current)

    numerical = tips_numerical_grade(final_grade)

    # OUTPUT
    print("\n----- GRADE REPORT -----")
    print(f"Name: {name}")
    print(f"Prelim Grade: {prelim_grade:.2f}")
    print(f"Midterm Grade: {midterm_grade:.2f}")
    print(f"Final Grade: {final_grade:.2f}")
    print(f"T.I.P.s Numerical Grade: {numerical}")
```
